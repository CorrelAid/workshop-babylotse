```python
import spacy
from collections import Counter, defaultdict

import data

nlp = spacy.load("de_dep_news_trf")
```

```python
df = data.load_data("Umfrage DE submissions 2024-07-31 00_25.csv")
df = df[df.Stadtteil != "SPAM"]
```

```python
df.head()
```

```python
for i in nlp.pipe(df["Positive"]):
    for noun_chunk in i.noun_chunks:
            print(noun_chunk.root.text, noun_chunk.root.head.text, noun_chunk.root.dep_)
    
```

```python
adj_list = []
adj_noun_dict = defaultdict(list)
adj_noun_list = []
head_pos_list = []

for i in nlp.pipe(df["Positive"]):
    for token in i:
        if token.pos_ == "ADJ":
            adj_list.append(token.lemma_.lower())
            adj_noun_dict[token.lemma_.lower()].append(token.head.lemma_.lower())
            adj_noun_list.append("{} {}".format(token.text.lower(), token.head))
            head_pos_list.append(token.head.pos_)


Counter(adj_list)
```
